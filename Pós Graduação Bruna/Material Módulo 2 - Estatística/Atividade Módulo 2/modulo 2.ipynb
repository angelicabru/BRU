Utilizando o pacote statistics:

```python
import statistics
```

Cálculo de média:

```python
media = statistics.mean(df['Preco'])
print('Média preço: ', media)
```

Cálculo de mediana:

```python
mediana = statistics.median(df['Preco'])
print('Mediana preço: ', mediana)
```

Cálculo de variância:

```python
variancia = statistics.variance(df['Preco'])
print('Variância de preço: ', variancia)
```

Cálculo de desvio padrão:

```python
desvio_padrao = statistics.stdev(df['Preco'])
print('Desvio padrão de preço: ', desvio_padrao)
```

Cálculo de amplitude:

```python
amplitude = max(df['Preco']) - min(df['Preco'])
print('Amplitude: ', amplitude)
```

Cálculo de moda:

```python
moda = statistics.mode(df['Preco'])
print('A moda é: ', moda)
```

Agora vamos importar uma nova biblioteca para utilizar os gráficos:

```python
import matplotlib.pyplot as plt
```

Criando um histograma de preço:

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Preco'], bins=10, edgecolor='black')
plt.title('Histograma do Preço')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

Criando sub gráficos:

```python
#Criar subplot
fig, axs = plt.subplots(3, figsize = (4,5))

#Primeiro subplot
axs[0].hist(df['Preco'], bins = 6, edgecolor = 'black')
axs[0].set_title('Distribuição de Preços de Carros')
axs[0].set_xlabel('Preço Carros')
axs[0].set_ylabel('Frequência')

#Segundo subplot
axs[1].hist(df['Quilometragem'], bins = 6, color = 'darkgreen', edgecolor = 'black')
axs[1].set_title('Distribuição de Venda por Quilometragem')
axs[1].set_xlabel('Quilometragem')
axs[1].set_ylabel('Frequência')

#Terceiro subplot
axs[2].hist(df['Idade_Locatario'], bins = 6, color = 'purple', edgecolor = 'white')
axs[2].set_title('Distribuição de Venda por Idade')
axs[2].set_xlabel('Idade_Locatario')
axs[2].set_ylabel('Frequência')

plt.tight_layout()

plt.show()
```

Gráfico Scatter Plot

```python
plt.title('Relação entre Preço e Idade Locatário')
plt.xlabel('Preço do Carro')
plt.ylabel('Idade Locatário')

plt.scatter(df['Preco'], df['Idade_Locatario'])

plt.show()
```

Separando o dataframe com ou sem ar condicionado:
