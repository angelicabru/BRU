```python
#referências:
#https://towardsdatascience.com/exploratory-data-analysis-in-python-a-step-by-step-process-d0dfa6bf94ee

#Vamos explorar um banco de dados de vinhos!

import pandas as pd

from sklearn.datasets import load_wine
vinhos = load_wine()

vinhos_df = pd.DataFrame(data=vinhos.data, columns=vinhos.feature_names)
```

```python
#Vamos supor que você está abrindo uma loja de vinhos

import pandas as pd
import numpy as np

#Converte para um pandas dataframe

vinhos_df = vinhos_df = pd.DataFrame(data=np.c_[vinhos['data'],vinhos['target']],columns=vinhos['feature_names']+['target'])

vinhos_df.info() #Primeira visualização das informações e atributos no geral
```

```python
#Vamos visualizar 5 registros aleatórios

vinhos_df.head() #ou .tail
```

```python
#Temos dados faltantes?

(vinhos_df.isnull() | vinhos_df.empty | vinhos_df.isna()).sum()

#.isnull verifica dados nulos
#.empty verifica dados vazios
#.isna verifica valores não válidos
```

```python
#Temos dados duplicados?

vinhos_df.duplicated().sum()
```

```python
#Vamos visualizar as principais estatísticas descritivas desse banco de dados através do comando describe:

vinhos_df.describe()
```

```python
#Também conseguimos visualizar a descrição de somente um atributo. Ex: coluna alcohol:

vinhos_df.alcohol.describe()
```

```python
#Vamos contar os atributos da coluna target:

vinhos_df['target'].value_counts()

#Target é a vinícola, e quando rodamos o código com o comando value_counts, somamos todos os valores das mesmas vinícolas. Muito similar a uma tabela dinâmica do Excel, por exemplo.
```

```python
#E quantos % representa cada vinícola? Podemos visualizar esses valores incluindo no código o comando normalize=True.

vinhos_df['target'].value_counts(normalize=True)
```

```python
#Que tal visualizar esses dados em um gráfico de barras simples? Vamos plotar com o comando plot.

vinhos_df['target'].value_counts().plot(kind='bar')
```

```python
#Ou também visualizar alguns atributos da base de dados através de histogramas:

vinhos_df[['alcohol','magnesium','color_intensity','flavanoids']].hist()
```

```python
#Na biblioteca seaborn temos o comando pairplot, nós geramos a visualização de todos os atributos vs todos os atributos, ou seja, todos os cruzamentos e visualizações possíveis para essa base de dados.

import seaborn as sns

sns.pairplot(vinhos_df)
```

```python
#Vamos rodar novamente o pairplot, mas focando em 4 atributos somente:

sns.pairplot(vinhos_df[['alcohol','magnesium','color_intensity','flavanoids']])
```

```python
#Vamos plotar um gráfico boxplot para verificar se há diferença significativa entre teor alcóolico e a vinícola de onde vem o vinho. 

import matplotlib.pyplot as plt

#referências:
#https://seaborn.pydata.org/generated/seaborn.catplot.html

sns.catplot(x='target', y='alcohol', data=vinhos_df, kind='box', aspect=1.5)
plt.title('Boxplot for target vs alcohol')
plt.show()

#Após rodar o código, é possível verificar que os vinhos da vinícola 1 tem menos teor alcóolico e o da vinícola 0 tem mais álcool.
```

```python
#Agora vamos plotar um histograma que mostre o teor alcóolico por vinícola.

import matplotlib.pyplot as plt

#Alcohol variable histograms

x1 = vinhos_df.loc[vinhos_df.target==0, 'alcohol']
x2 = vinhos_df.loc[vinhos_df.target==1, 'alcohol']
x3 = vinhos_df.loc[vinhos_df.target==2, 'alcohol']

kwargs = dict(alpha=0.3, bins=25)

plt.hist(x1, **kwargs, color='g', label='Tipo 0')
plt.hist(x2, **kwargs, color='b', label='Tipo 1')
plt.hist(x3, **kwargs, color='r', label='Tipo 2')
plt.gca().set(title='Teor alcóolico por vinícola', ylabel='Frequência')
plt.legend()
```

```python
#E também vamos plotar um histograma que mostre o teor de magnésio por vinícola.

import matplotlib.pyplot as plt

#Alcohol variable histograms

x1 = vinhos_df.loc[vinhos_df.target==0, 'magnesium']
x2 = vinhos_df.loc[vinhos_df.target==1, 'magnesium']
x3 = vinhos_df.loc[vinhos_df.target==2, 'magnesium']

kwargs = dict(alpha=0.3, bins=25)

plt.hist(x1, **kwargs, color='g', label='Tipo 0')
plt.hist(x2, **kwargs, color='b', label='Tipo 1')
plt.hist(x3, **kwargs, color='r', label='Tipo 2')
plt.gca().set(title='Teor de magnésio por vinícola', ylabel='Frequência')
plt.legend()
```

```python
#Dominar visualizações tem valor!
#Como mostrar a média?

x1 = vinhos_df.loc[vinhos_df.target==0, 'alcohol']
x2 = vinhos_df.loc[vinhos_df.target==1, 'alcohol']
x3 = vinhos_df.loc[vinhos_df.target==2, 'alcohol']

kwargs = dict(alpha=0.3, bins=25)

plt.hist(x1, **kwargs, color='g', label='Tipo 0' + str('{:6.2f}'.format(x1.std())))
plt.hist(x2, **kwargs, color='b', label='Tipo 1' + str('{:6.2f}'.format(x2.std())))
plt.hist(x3, **kwargs, color='r', label='Tipo 2' + str('{:6.2f}'.format(x3.std())))
plt.gca().set(title='Frequência de alcóol por tipo de vinho', ylabel='Frequência')

#O código abaixo vai plotar uma linha exatamente na média alcóolica de cada vinícola:

plt.axvline(x1.mean(), color='g', linestyle='dashed', linewidth=1)
plt.axvline(x2.mean(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(x3.mean(), color='r', linestyle='dashed', linewidth=1)

plt.legend()
```

```python
#Mesma lógica, mas agora vamos visualizar o teor de magnésio com relação a cada vinícola:

x1 = vinhos_df.loc[vinhos_df.target==0, 'magnesium']
x2 = vinhos_df.loc[vinhos_df.target==1, 'magnesium']
x3 = vinhos_df.loc[vinhos_df.target==2, 'magnesium']

kwargs = dict(alpha=0.3, bins=25)

plt.hist(x1, **kwargs, color='g', label='Tipo 0' + str('{:6.2f}'.format(x1.std())))
plt.hist(x2, **kwargs, color='b', label='Tipo 1' + str('{:6.2f}'.format(x2.std())))
plt.hist(x3, **kwargs, color='r', label='Tipo 2' + str('{:6.2f}'.format(x3.std())))
plt.gca().set(title='Frequência de magnésio por tipo de vinho', ylabel='Frequência')

#O código abaixo vai plotar uma linha exatamente na média de magnésio de cada vinícola:

plt.axvline(x1.mean(), color='g', linestyle='dashed', linewidth=1)
plt.axvline(x2.mean(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(x3.mean(), color='r', linestyle='dashed', linewidth=1)
```

```python
#Vamos verificar a correlação de alguns valores com o comando .corr:

vinhos_df[['alcohol','magnesium','color_intensity']].corr()
```

```python
#A partir disso, vamos plotar um gráfico de dispersão para visualizar melhor essa correlação:

import seaborn as sns
import matplotlib.pyplot as plt 

sns.scatterplot(x='color_intensity', y='alcohol', hue='target', data=vinhos_df, palette='Dark2', s=80)
plt.title('Relação entre color_intensity, alcohol e target')
plt.show()

#'hue' é onde vai ocorrer a diferença de cor, e aqui definimos que será o tipo de vinho/vinícola.
```

```python
#Vamos gerar um pairplot diferenciando/destacando os valores pelo target/vinícola.

import seaborn as sns

df = vinhos_df[['alcohol','magnesium','color_intensity','target']]
palette = sns.color_palette('bright')
sns.pairplot(df, palette=palette, hue='target')
```

```python
#É fácil verificar quais variáveis são fortemente correlacionadas utilizando o comando .corr:

corrmat = vinhos_df.corr()

hm = sns.heatmap(corrmat,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 5},
                 yticklabels=df.columns,
                 xticklabels=df.columns,
                 cmap='Spectral_r')
plt.show()
```
