### CAPÍTULO 1

**1. Ciclo de vida de projetos de dados**

Entendimento do problema de negócio e entendimento dos dados interagem!



**Exemplos de problemas de negócio**

-Análise de Churn (perda de clientes);

-Segmentação de mercado;

-Previsão de vendas;

-Detecção de fraude;

-Otimização de preços;

-Recomendação de produtos;

-Análise de sentimento;

-Gerenciamento de inventário;

-Aumento da eficiência operacional;

-Análise de risco.

**Entendimento do problema de negócio**

Em um mundo real e ideal...

-As metas e objetivos de negócio estão claros e alinhados;

-Os objetivos de negócio são mensuráveis e realistas;

-Os dados existem e são de qualidade.

**Entendimento dos dados**

-Identificar fontes de dados;

-Bancos de dados internos;

-APIs externas;

-Conjuntos de dados públicos;

-Avaliar a disponibilidade e acessibilidade dos dados;

-Compreender formatos e estruturas dos dados.

#### Desafio 1: Problema é ambíguo

**Exemplo:** Um cliente deseja aumentar a retenção de clientes:

-Quais aspectos da retenção são mais importantes? 

-Quais ações específicas devem ser tomadas? 

Solução: trabalhar com o cliente para definir claramente os objetivos, metas e expectativas.

*Como os dados ajudam:*

-Analisar os dados pode revelar padrões e insights que ajudam a clarificar o problema. 

-Quais variáveis estão mais relacionadas com a retenção de clientes?

#### Desafio 2: Falta alinhamento com necessidade de negócio

**Exemplo:** Uma empresa quer implementar um modelo preditivo para vendas, mas o porblema real é a baixa taxa de conversão no funil de vendas.

Solução: identificar o problema subjacente e alinhar o projeto com as necessidades reais de negócios.

#### Desafio 3: Objetivos de negócio não mensuráveis

**Exemplo:** Um cliente deseja melhorar a "satisfação do cliente" sem definir métricas claras ou formas de medir essa satisfação.

Solução: trabalhar com o cliente para definir KPIs e métricas específicas para mensurar a satisfação.

*Como os dados ajudam:*

-A análise dos dados pode ajudar a definir métricas e KPIs apropriados. 

-Pode identificar quais variáveis estão mais associadas à satisfação, possibilitando a definição de métricas.

#### Desafio 4: Expectativas irrealistas

**Exemplo:** A expectativa de que um modelo de aprendizado de máquina pode resolver todos os problemas de um processo complexo sem considerar limitações e restrições. 

Solução: educar os stakeholders sobre o que é viável e quais são as limitações dos modelos de dados.

*Como os dados ajudam:*

Entendimento dos dados ajuda a identificar limitações dos dados e dos modelos que podem ser utilizados.

#### Desafio 5: Dados de baixa qualidade

**Exemplo:** O objetivo é prever a demanda de um produto, mas os dados históricos são incompletos ou têm baixa qualidade. 

*Como os dados ajudam:*

-Avaliar a qualidade dos dados disponíveis.

-Identificar possíveis fontes adicionais de dados.



**2. Técnicas de análise exploratória de dados**

**O que é Análise Exploratória de Dados?**

Processo que investiga os seus dados para descobrir: Padrões, anomalias, relações e tendências (por meio de estatísticas e métodos visuais).

*Por que é importante?*

-Útil no início de um projeto de dados

-Verificar a viabilidade de projetos de dados

-Entende a estrutura dos dados antes de aplicar métodos de aprendizado de máquina mais sofisticados

-Refina e ajusta as hipóteses iniciais

-Traz insights sobre a estrutura dos dados

-Te dá confiança sobre a qualidade dos dados

**Técnicas básicas de AED (Análise Exploratória de Dados)**



a) Distribuição dos dados

-Entender a variabilidade dos dados;

-Dados são normais? Assimétricos?;

-Estatísticas que sumarizam os dados;

-Histogramas e Box Plots;

-Comparar grupos.

b) Dados faltantes

-Algoritmos de aprendizado de máquina são sensíveis a dados faltantes;

-Sintoma de qualidade dos dados;

-Imputação;

-Deleção;

-Entendimento das razões pelos quais faltam dados.

c) Qualidade dos dados

identificar erros e inconsistências nos dados.

d) Exceções/Anomalias

-Revelam dados que fogem do padrão esperado;

-Oportunidades inesperadas;

-Remover? Transformar?

e) Correlação

Revelam relações e potenciais dependências. Qual a direção e força das relações?

f) Padrões em geral

g) Tipos de dados

Tipo de técnica muda de acordo com o tipo de dados (numérico, categórico, ordinal, data, hora etc.)

h) Visualização dos dados
