**1. Ciclo de vida de projetos de dados**

Entendimento do problema de negcio e entendimento dos dados interagem!



**Exemplos de problemas de negcio**

-Anlise de Churn (perda de clientes);

-Segmentao de mercado;

-Previso de vendas;

-Deteco de fraude;

-Otimizao de preos;

-Recomendao de produtos;

-Anlise de sentimento;

-Gerenciamento de inventrio;

-Aumento da eficincia operacional;

-Anlise de risco.

**Entendimento do problema de negcio**

Em um mundo real e ideal...

-As metas e objetivos de negcio esto claros e alinhados;

-Os objetivos de negcio so mensurveis e realistas;

-Os dados existem e so de qualidade.

**Entendimento dos dados**

-Identificar fontes de dados;

-Bancos de dados internos;

-APIs externas;

-Conjuntos de dados pblicos;

-Avaliar a disponibilidade e acessibilidade dos dados;

-Compreender formatos e estruturas dos dados.

#### Desafio 1: Problema  ambguo

**Exemplo:** Um cliente deseja aumentar a reteno de clientes:

-Quais aspectos da reteno so mais importantes? 

-Quais aes especficas devem ser tomadas? 

Soluo: trabalhar com o cliente para definir claramente os objetivos, metas e expectativas.

*Como os dados ajudam:*

-Analisar os dados pode revelar padres e insights que ajudam a clarificar o problema. 

-Quais variveis esto mais relacionadas com a reteno de clientes?

#### Desafio 2: Falta alinhamento com necessidade de negcio

**Exemplo:** Uma empresa quer implementar um modelo preditivo para vendas, mas o porblema real  a baixa taxa de converso no funil de vendas.

Soluo: identificar o problema subjacente e alinhar o projeto com as necessidades reais de negcios.

#### Desafio 3: Objetivos de negcio no mensurveis

**Exemplo:** Um cliente deseja melhorar a "satisfao do cliente" sem definir mtricas claras ou formas de medir essa satisfao.

Soluo: trabalhar com o cliente para definir KPIs e mtricas especficas para mensurar a satisfao.

*Como os dados ajudam:*

-A anlise dos dados pode ajudar a definir mtricas e KPIs apropriados. 

-Pode identificar quais variveis esto mais associadas  satisfao, possibilitando a definio de mtricas.

#### Desafio 4: Expectativas irrealistas

**Exemplo:** A expectativa de que um modelo de aprendizado de mquina pode resolver todos os problemas de um processo complexo sem considerar limitaes e restries. 

Soluo: educar os stakeholders sobre o que  vivel e quais so as limitaes dos modelos de dados.

*Como os dados ajudam:*

Entendimento dos dados ajuda a identificar limitaes dos dados e dos modelos que podem ser utilizados.

#### Desafio 5: Dados de baixa qualidade

**Exemplo:** O objetivo  prever a demanda de um produto, mas os dados histricos so incompletos ou tm baixa qualidade. 

*Como os dados ajudam:*

-Avaliar a qualidade dos dados disponveis.

-Identificar possveis fontes adicionais de dados.
